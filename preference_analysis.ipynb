```python
import pandas as pd

df = pd.read_excel("pcp_outcome3.xlsx")
df = df.iloc[:, 2:8]
df = df.rename(columns={'以下app中您经常使用的是（必填）': 'app_name', '以下app中您用来看视频的是（必填）': 'app_video', '您多久看一次娱乐类视频（最接近的次数）（必填）': 'frequency', '您对看功夫类的视频感兴趣吗（必填）': 'kungfu_interest', '关于功夫，您更想看关于它的哪方面的视频（必填）': 'kungfu_aspect', '您听说过李小龙吗（必填）': 'Bruce_Lee'})

df['app_video'] = df['app_video'].str.lower()

df['app_name'] = df['app_name'].str.split(', ')
df['app_video'] = df['app_video'].str.split(', ')
df['kungfu_aspect'] = df['kungfu_aspect'].str.split(', ')

df
```

```python
from itertools import chain
from collections import Counter

app_name_count = pd.Series(Counter(chain.from_iterable(df['app_name'])))
app_video_count = pd.Series(Counter(chain.from_iterable(df['app_video'])))
kungfu_aspect_count = pd.Series(Counter(chain.from_iterable(df['kungfu_aspect'])))

print("------app_name_count-------\n")
print(app_name_count)
print("\n------app_video_count-------\n")
print(app_video_count)
print("\n------kungfu_aspect_count-------\n")
print(kungfu_aspect_count)
```

```python
#normalize the data
def normalize_data(df):
    
    total = 0
    new_df = [0] * len(df)
    for i in range(len(df)):
        total += df[i]
        
    for j in range(len(df)):
        new_df[j] = df[j] / total
    
    
    return new_df
```

```python
app_name_df = pd.DataFrame(app_name_count).rename(columns = {0: "Count"})
app_video_df = pd.DataFrame(app_video_count).rename(columns = {0: "Count"})
kungfu_aspect_df = pd.DataFrame(kungfu_aspect_count).rename(columns = {0: "Count"})

kungfu_aspect_df = kungfu_aspect_df.sort_values(by=['Count'], ascending=False)
```

```python
#start visualizing the data
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


#this function reads in labels and the data, and produce a pie chart with the largest part indented out
def customized_plot_pie(labels, sizes, title):
    
    font = fm.FontProperties(fname=r'C:\Users\hp\Downloads\hpsimplifiedhans-regular.ttf')
    
    explode = [0] * len(labels)
    explode[0] = 0.1
    
    fig1, ax1 = plt.subplots()
    patches, texts, autotexts = ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    ax1.set_title(title)
    
    plt.setp(autotexts, fontproperties=font)
    plt.setp(texts, fontproperties=font)

    plt.show()
```

```python
customized_plot_pie(kungfu_aspect_df.index, kungfu_aspect_df["Count"], "What Audience Like to Watch About Kung Fu")
```

```python
neither_audience =  df[df["kungfu_interest"] == "都行 Neither"]
neither_aspect_count = pd.Series(Counter(chain.from_iterable(neither_audience['kungfu_aspect'])))
neither_aspect_df = pd.DataFrame(neither_aspect_count).rename(columns = {0: "Count"})

neither_aspect_df = neither_aspect_df.sort_values(by=['Count'], ascending=False)
customized_plot_pie(neither_aspect_df.index, neither_aspect_df["Count"], "What Audience Who Chose Neither Like to Watch About Kung Fu")
```

```python
import numpy as np
def customized_plot_bar(label, data):
    y_pos = np.arange(len(label))
    fig, ax = plt.subplots()
    
    font = fm.FontProperties(fname=r'C:\Users\hp\Downloads\hpsimplifiedhans-regular.ttf')
    
    hbars = ax.barh(y_pos, data, align='center')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(label, fontproperties=font)
    ax.invert_yaxis()
    ax.set_xlabel('Count')
    ax.set_title('Number of People Using Each App')
    
    ax.set_xlim(right = 50)
    
    plt.show()
    
```

```python
def customized_vertical_bar(label, data, title):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(label, data)
    
    font = fm.FontProperties(fname=r'C:\Users\hp\Downloads\hpsimplifiedhans-regular.ttf')
    plt.xticks(rotation=30, fontproperties=font)
    
    ax.set_title(title)
    
    plt.show()
```

```python
def customized_stacked_bar(label, legend1, legend2, data1, data2, title):    
    fig, ax = plt.subplots()
    
    font = fm.FontProperties(fname=r'C:\Users\hp\Downloads\hpsimplifiedhans-regular.ttf')
    plt.xticks(fontproperties=font)
    
    plt.title(title)
    
    plt.bar(label, data1)
    plt.bar(label, data2, bottom=data1)
    
    ax.legend([legend1, legend2], bbox_to_anchor=(1.05, 1), loc='upper left', prop=font);
    plt.show()
```
